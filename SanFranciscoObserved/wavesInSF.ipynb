```python
def add_datetime_column(df):                #Given df with seperate year, month, day, hour, min and second construct a date column
    year = list(df['#YY'])
    month = list(df['MM'])
    day = list(df['DD'])
    hour = list(df['hh'])
    minute = list(df['mm'])
    date_column = ['date']
    for i in range(1, len(df)):
        date_column.append(datetime(year = int(year[i]), month= int(month[i]), day = int(day[i]), hour = int(hour[i]), minute = int(minute[i])))
    df['Date'] = date_column
    return df

def drop_unnecessary_columns(df):
    cols_ro_remove = ['#YY_x', 'MM_x', 'DD_x', 'hh_x', 'mm_x','WVHT_x',
       'DPD_x', 'APD_x', 'TIDE_x', 'MWD   PRES_x', 'ATMP_x', 'DEWP_x', 'VIS_x', '#YY_y', 'MM_y', 'DD_y', 'hh_y', 'mm_y', 'ATMP_y', 'DEWP_y', 'VIS_y', 'TIDE_y']
    thinned_df = df.drop(columns = cols_ro_remove)
    return(thinned_df.head(10))
def remove_nan_columns(df):
    for column in df.columns:
        sd_column = np.std(df[column][1:])
        print(sd_column)

sf_waves_2021 = add_datetime_column(sf_waves_2021)
sf_waves_2022 = add_datetime_column(sf_waves_2022)
sf_winds_2021 = add_datetime_column(sf_winds_2021)
sf_winds_2022 = add_datetime_column(sf_winds_2022)
sf_data_2021 = sf_winds_2021.merge(sf_waves_2021, left_on='Date', right_on='Date')
sf_data_2022 = sf_winds_2022.merge(sf_waves_2022, left_on='Date', right_on='Date')
drop_unnecessary_columns(sf_data_2021)
```

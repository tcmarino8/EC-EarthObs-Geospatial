```python
### Ok so we gotta figure out how to get each ticket a coordinate! 
## Functino Works if data set contains a dictionary as geom: a Point data tyoe
def add_coords(dataset, geom_column):
    coordinates = []
    for i in range(len(dataset)):
        if str(dataset[geom_column][i]) == 'nan':
            coordinates.append([0,0])
        else:
            coordinates.append(dataset[geom_column][i].get('coordinates'))
    dataset['coordinates'] = coordinates

add_coords(dontParkSF, 'the_geom')
dontParkSF['coordinates']
```

```python
coords = dontParkSF['coordinates']
sum_x = 0
sum_y = 0
count = 0
for i in range(len(coords)):
    if coords[i][0] == 0:
        count += 0
        sum_x += 0
        sum_y += 0
    else:
        count += 1
        sum_x += int(coords[i][0])
        sum_y += int(coords[i][1])
avg_coord = [sum_x/count, sum_y/count]
```
